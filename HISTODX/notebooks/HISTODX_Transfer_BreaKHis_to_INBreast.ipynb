```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = CURRENT
if not (PROJECT_ROOT / "HISTODX_TORCH").exists():
    for parent in [CURRENT, *CURRENT.parents]:
        if (parent / "HISTODX_TORCH").exists():
            PROJECT_ROOT = parent
            break

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA:", torch.version.cuda)
print("GPU:", torch.cuda.is_available())
```

```python
from HISTODX_TORCH import (
    make_run_dirs,
    run_histodx_breakhis_baseline,
    run_histodx_transfer_breakhis_to_inbreast,
)

BREAKHIS_BASE_PATH = PROJECT_ROOT / "BrestCancer Datasets" / "BreaKHis" / "BreaKHis_v1" / "BreaKHis_v1" / "histology_slides" / "breast"
INBREAST_CSV = PROJECT_ROOT / "BrestCancer Datasets" / "INBreast" / "INbreast" / "INbreast.csv"
INBREAST_DICOM_DIR = PROJECT_ROOT / "BrestCancer Datasets" / "INBreast" / "INbreast" / "AllDICOMs"
```

```python
source_run_dirs = make_run_dirs(run_prefix="histodx_breakhis_source")
source_results = run_histodx_breakhis_baseline(
    dataset_path=str(BREAKHIS_BASE_PATH),
    epochs=5,
    batch_size=32,
    run_dirs=source_run_dirs,
    balance=True,
)
print("Run fonte:", source_run_dirs["out_dir"])
```

```python
target_run_dirs = make_run_dirs(run_prefix="histodx_breakhis_to_inbreast_tl")
target_results = run_histodx_transfer_breakhis_to_inbreast(
    inbreast_csv_path=str(INBREAST_CSV),
    dicom_dir=str(INBREAST_DICOM_DIR),
    mode="multiclass",
    epochs=5,
    batch_size=16,
    run_dirs=target_run_dirs,
    balance=True,
    freeze_except_classifier=True,
)
target_results["metrics"]
```
