```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for parent in [CURRENT, *CURRENT.parents]:
    if (parent / "histology_datasets.py").exists() and (parent / "HISTODX").exists():
        PROJECT_ROOT = parent
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Nao encontrei a raiz do projeto.")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA disponivel:", torch.cuda.is_available())
```

```python
from HISTODX import make_run_dirs, run_histodx_bach_baseline, run_histodx_bracs_baseline

DATASET_NAME = "bracs"  # troque para "bach" quando quiser
DATASET_PATHS = {
    "bracs": PROJECT_ROOT / "BrestCancer Datasets" / "BRACS",
    "bach": PROJECT_ROOT / "BrestCancer Datasets" / "BACH",
}

EPOCHS = 10
BATCH_SIZE = 16
IMG_SIZE = 224
LEARNING_RATE = 1e-4
BALANCE = True
PRETRAINED = False

DATASET_PATH = DATASET_PATHS[DATASET_NAME]
print("Dataset:", DATASET_NAME, DATASET_PATH)
```

```python
run_dirs = make_run_dirs(run_prefix=f"histodx_{DATASET_NAME}_baseline")

if DATASET_NAME == "bracs":
    results = run_histodx_bracs_baseline(
        dataset_path=str(DATASET_PATH),
        img_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        learning_rate=LEARNING_RATE,
        run_dirs=run_dirs,
        balance=BALANCE,
        pretrained=PRETRAINED,
        seed=42,
    )
else:
    results = run_histodx_bach_baseline(
        dataset_path=str(DATASET_PATH),
        img_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        learning_rate=LEARNING_RATE,
        run_dirs=run_dirs,
        balance=BALANCE,
        pretrained=PRETRAINED,
        val_fraction=0.15,
        test_fraction=0.15,
        seed=42,
    )

print("Run salvo em:", results["run_dirs"]["out_dir"])
results["metrics"]
```
