```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for parent in [CURRENT, *CURRENT.parents]:
    if (parent / "histology_datasets.py").exists() and (parent / "HISTODX").exists():
        PROJECT_ROOT = parent
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Nao encontrei a raiz do projeto.")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA disponivel:", torch.cuda.is_available())
```

```python
from HISTODX import make_run_dirs, run_histodx_bach_baseline

BACH_PATH = PROJECT_ROOT / "BrestCancer Datasets" / "BACH"
EPOCHS = 200
BATCH_SIZE = 16
IMG_SIZE = 224
LEARNING_RATE = 1e-4
BALANCE = True
PRETRAINED = False
VAL_FRACTION = 0.15
TEST_FRACTION = 0.15
SEED = 42

print("Dataset:", BACH_PATH)
```

```python
run_dirs = make_run_dirs(run_prefix="histodx_bach_baseline")
results = run_histodx_bach_baseline(
    dataset_path=str(BACH_PATH),
    img_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    learning_rate=LEARNING_RATE,
    run_dirs=run_dirs,
    balance=BALANCE,
    pretrained=PRETRAINED,
    val_fraction=VAL_FRACTION,
    test_fraction=TEST_FRACTION,
    seed=SEED,
)

print("Run salvo em:", results["run_dirs"]["out_dir"])
results["metrics"]
```
