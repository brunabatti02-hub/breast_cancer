```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for parent in [CURRENT, *CURRENT.parents]:
    if (parent / "histology_datasets.py").exists() and (parent / "SEConformer").exists():
        PROJECT_ROOT = parent
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Nao encontrei a raiz do projeto.")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA disponivel:", torch.cuda.is_available())
```

```python
from SEConformer import make_run_dirs, train_bach_baseline, train_bracs_baseline

DATASET_NAME = "bracs"  # troque para "bach" quando quiser
DATASET_PATHS = {
    "bracs": PROJECT_ROOT / "BrestCancer Datasets" / "BRACS",
    "bach": PROJECT_ROOT / "BrestCancer Datasets" / "BACH",
}

EPOCHS = 10
BATCH_SIZE = 16
IMG_SIZE = 224
LR = 1e-4
BALANCE = True

DATASET_PATH = DATASET_PATHS[DATASET_NAME]
print("Dataset:", DATASET_NAME, DATASET_PATH)
```

```python
run_dirs = make_run_dirs(run_prefix=f"seconformer_{DATASET_NAME}_baseline")

if DATASET_NAME == "bracs":
    model, run_dirs, metrics = train_bracs_baseline(
        dataset_path=str(DATASET_PATH),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        img_size=IMG_SIZE,
        lr=LR,
        run_dirs=run_dirs,
        balance=BALANCE,
    )
else:
    model, run_dirs, metrics = train_bach_baseline(
        dataset_path=str(DATASET_PATH),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        img_size=IMG_SIZE,
        lr=LR,
        run_dirs=run_dirs,
        balance=BALANCE,
        val_fraction=0.15,
        test_fraction=0.15,
        seed=42,
    )

print("Run salvo em:", run_dirs["out_dir"])
metrics
```
