```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for parent in [CURRENT, *CURRENT.parents]:
    if (parent / "histology_datasets.py").exists() and (parent / "SEConformer").exists():
        PROJECT_ROOT = parent
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Nao encontrei a raiz do projeto.")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA disponivel:", torch.cuda.is_available())
```

```python
from SEConformer import make_run_dirs, train_bracs_baseline

BRACS_PATH = PROJECT_ROOT / "BrestCancer Datasets" / "BRACS"
EPOCHS = 30
BATCH_SIZE = 16
IMG_SIZE = 224
LR = 1e-4
BALANCE = True
USE_MODERATE_AUGMENTATION = True
EPOCH_SIZE_MULTIPLIER = 1.25

print("Dataset:", BRACS_PATH)
```

```python
run_dirs = make_run_dirs(run_prefix="seconformer_bracs_baseline")
model, run_dirs, metrics = train_bracs_baseline(
    dataset_path=str(BRACS_PATH),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    img_size=IMG_SIZE,
    lr=LR,
    run_dirs=run_dirs,
    balance=BALANCE,
    use_moderate_augmentation=USE_MODERATE_AUGMENTATION,
    epoch_size_multiplier=EPOCH_SIZE_MULTIPLIER,
)

print("Run salvo em:", run_dirs["out_dir"])
metrics
```
