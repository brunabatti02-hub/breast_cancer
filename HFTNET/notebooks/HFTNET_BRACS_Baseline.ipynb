```python
import sys
from pathlib import Path

CURRENT = Path.cwd().resolve()
PROJECT_ROOT = None
for parent in [CURRENT, *CURRENT.parents]:
    if (parent / "histology_datasets.py").exists() and (parent / "HFTNET").exists():
        PROJECT_ROOT = parent
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError("Nao encontrei a raiz do projeto.")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import torch
print("Projeto:", PROJECT_ROOT)
print("Torch:", torch.__version__)
print("CUDA disponivel:", torch.cuda.is_available())
```

```python
from HFTNET import make_run_dirs, run_bracs_baseline

BRACS_PATH = PROJECT_ROOT / "BrestCancer Datasets" / "BRACS"
EPOCHS = 10
BATCH_SIZE = 8
IMG_SIZE = 224
LR = 1e-4
BALANCE = True
PRETRAINED = False

print("Dataset:", BRACS_PATH)
```

```python
run_dirs = make_run_dirs(run_prefix="hftnet_bracs_baseline")
model, history, best_outputs, run_dirs = run_bracs_baseline(
    dataset_path=str(BRACS_PATH),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    img_size=IMG_SIZE,
    lr=LR,
    run_dirs=run_dirs,
    balance=BALANCE,
    pretrained=PRETRAINED,
)

print("Run salvo em:", run_dirs["out_dir"])
best_outputs[3]
```
